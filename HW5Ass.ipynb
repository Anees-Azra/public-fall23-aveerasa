```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/10s0013A.csv')
```

```python
df.info()
```

```python
column_names=[
 'State',
 'Post office abbreviation',
 '2-digit ANSI code',
 '1960_RANK',
 '1970_RANK',
 '1980_RANK',
 '1990_RANK',
 '2000_RANK',
 '2008_RANK',
 '1960_1970_PERCENT_CHANGE',
 '1970_1980_PERCENT_CHANGE',
 '1980_1990_PERCENT_CHANGE',
 '1980_1990_PERCENT_CHANGE.1',
 '1990_2000_PERCENT_CHANGE',
 '2000_2008_PERCENT_CHANGE',
 '1960_Population',
 '1970_Population',
 '1980_Population',
 '1990_Population',
 '2000_Population',
 '2008_Population'
]
df.columns = column_names
```

```python
df.info()
```

```python
df['1960_Population'] = df['1960_Population'].str.replace(',', '').astype(float)
df['1970_Population'] = df['1970_Population'].str.replace(',', '').astype(float)
df['1980_Population'] = df['1980_Population'].str.replace(',', '').astype(float)
df['1990_Population'] = df['1990_Population'].str.replace(',', '').astype(float)
df['2000_Population'] = df['2000_Population'].str.replace(',', '').astype(float)
df['2008_Population'] = df['2008_Population'].str.replace(',', '').astype(float)
```

```python
df.info()
```

```python
df.fillna(0, inplace=True)
```

```python
df.isna().value_counts()
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[df['State'].str.strip() != 'District of Columbia']
```

```python
df.dropna()
```

```python
columns_to_plot = ['1990_Population', '2000_Population', '2008_Population']
```

```python
df_plot.describe()
```

```python
print(df['1990_Population'].median(),df['2000_Population'].median(), df['2008_Population'].median() )
```

```python
# Create a box plot using Seaborn
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.boxplot(data=df[columns_to_plot],whis=[5, 95], showfliers=True)  # Adjust whis as needed

# Customize the plot
plt.title("Population Distribution Over the Years")
plt.xlabel("Year")
plt.ylabel("Population")

# Adjust the y-axis limits to make the boxes appear large
plt.ylim(0, 5000)  # Adjust the upper limit as needed

# Show the plot
plt.show()
```

```python
population_density_1990 = df['1990_Population']

# Create a figure with two subplots for the eCDF and histogram
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Create the eCDF
sns.ecdfplot(data=population_density_1990, ax=ax1)
ax1.set_title("Empirical CDF of Population Density For Year 1990")
ax1.set_xlabel("Population Density")
ax1.set_ylabel("eCDF")

# Create the histogram
sns.histplot(data=population_density_1990, bins=30, kde=True, ax=ax2)
ax2.set_title("Histogram of Population Density For Year 1990")
ax2.set_xlabel("Population Density")
ax2.set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
